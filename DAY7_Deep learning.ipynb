```python
import numpy as np
```

```python
# 입력층

input_data = np.array([[2,3], [5,1]])
print(input_data)

x= input_data.reshape(-1) # 행렬 일자로 펴주기
print(x)
```

```python
#가중치 및 편향(임의로 설정함)

w1= np.array([2,1,-3,3]) #가중치
w2= np.array([1,-3,1,3]) #가중치
b1 = 3
b2 = 3
```

```python
# 가중합

W = np.array([w1,w2])
print(W)

b = np.array([b1,b2])
print(b)

weight_sum = np.dot(W,x) + b # 내적값 구하기
print(weight_sum)
```

```python
#출력층

res = 1/(1+np.exp(-weight_sum)) #시그모이드 함수 적용
print(res)
```

## 2. 딥러닝 분류

```python
import numpy as np
import pandas as pd
import tensorflow as tf

np.random.seed(0)
tf.random.set_seed(0)
```

```python
df = pd.read_csv('wine_data.csv')
```

```python
df.head()
```

```python
#피처, 타겟 데이터 지정
```

```python
X = df.iloc[:,0:13]
y = df.iloc[:,13]
```

```python
X
```

```python
# 피처 데이터 차원 확인
```

```python
print(X.shape)
```

```python
#타겟 데이터 종류 확인
```

```python
print(set(y))
```

```python
# 타겟 데이터 원-핫 인코딩
```

```python
from tensorflow.keras.utils import to_categorical
y_hot = to_categorical(y) 
print(y_hot)
```

```python
# 트레이닝/테스트 데이터 분할
from sklearn.model_selection import train_test_split
X_tn, X_te, y_tn, y_te=train_test_split(X, y_hot, random_state=0)
```

```python
#신경망 생성
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Activation

n_feat = X_tn.shape[1]
n_class = len(set(y))
epo= 30

model = Sequential()
model.add(Dense(20, input_dim=n_feat))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense(n_class))
model.add(Activation('softmax'))
```

```python
#신경망 모형 구조 확인
```

```python
model.summary()
```

```python
#모형 컴파일
```

```python
model.compile(loss ='categorical_crossentropy',
             optimizer ='adam',
             metrics=['accuracy'])
```

```python
# 신경망 학습
hist = model.fit(X_tn, y_tn, epochs=epo, batch_size=5)
```

```python
#트레이닝 데이터 평가
print(model.evaluate(X_tn, y_tn)[1])
```

```python
#테스트 데이터 평가
print(model.evaluate(X_te, y_te)[1])
```

```python
import numpy as np
epoch = np.arange(1,epo+1)
print(epoch)
```

```python
accuracy = hist.history['accuracy']
print(accuracy)
```

```python
loss = hist.history['loss']
print(loss)
```

```python
# 정확도 학습 그래프
import matplotlib.pyplot as plt

plt.plot(epoch, accuracy, label='accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

```python
#손실 그래프
import matplotlib.pyplot as plt

plt.plot(epoch, loss, 'r', label='loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf

np.random.seed(0)
tf.random.set_seed(0)
```

```python
df = pd.read_csv('boston_house_prices.csv')
```

```python
df.head()
```

```python
#피처, 타겟 데이터 지정
X = df.iloc[:,0:13]
y = df.iloc[:,13]
```

```python
X
```

```python
#피처 데이터 차원 확인
print(X.shape)
```

```python
#타겟 데이터 종류 확인
#여기서는 y 자체 값을 구하는 (집값)임
print(set(y))
```

```python
# 트레이닝/테스트 데이터 분할
from sklearn.model_selection import train_test_split
X_tn, X_te, y_tn, y_te=train_test_split(X, y, random_state=0)
#여기서는 y_hot 이 아니라 y (집값)
```

```python
#신경망 생성
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

n_feat = X_tn.shape[1]
epo= 30

model = Sequential()
model.add(Dense(20, input_dim=n_feat, activation='relu'))
model.add(Dense(1))
model.summary()
```

```python
# 모형 컴파일
model.compile(loss ='mean_squared_error',
             optimizer ='adam',
             metrics=['mean_squared_error'])
```

```python
# 신경망 학습
hist= model.fit(X_tn, y_tn, epochs=epo, batch_size =5)
```

```python
#트레이닝 데이터 평가
print(model.evaluate(X_tn, y_tn)[1])
```

```python
# 테스트 데이터 평가
print(model.evaluate(X_te, y_te)[1])
```
