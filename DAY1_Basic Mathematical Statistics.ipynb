### 3. 팩토리얼 

```python
# 팩토리얼 함수화 하기!
def factorial(n):
    res = 1
    for i in range(1, n+1) :
        res *= i
    return res
```

```python
factorial(5)
```

### 4. 이항함수

```python
# 만들어둔 factorial (n) 함수 삿용
factorial(4)/(factorial(2) * factorial(4-2))
```

```python
def binomial(n,x) :
    res = factorial(n)/(factorial(x) * factorial (n-x))
    return res
```

```python
binomial(4,2)
```
